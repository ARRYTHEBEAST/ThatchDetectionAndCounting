```python
import os
import glob
import shutil
from ultralytics import YOLO

save_path = "/kaggle/working/yolov8-mskf"

# Function to perform detection
def run_detection(weights, source, conf, save_txt, save_conf, augment, name):
    model = YOLO(weights)  # Initialize the YOLO model
    model.predict(
        source=source,
        #conf_thres=conf,
        save_txt=save_txt,
        save_conf=save_conf,
        #nosave=nosave,
        augment=augment,
        name=name
    )

# Loop to run detection for each fold
for i in range(5):
    run_detection(
        weights=f"/kaggle/input/thatchdetectionweights/best_{i}.pt",
        #img_size=1536,
        source=f"/kaggle/input/thatchdetectionfolddata/data/fold_{i}/images",
        conf=0.1,
        save_txt=True,
        save_conf=True,
        #nosave=True,
        augment=True,
        name=f"yolov8-custom-different-augs-image-size-1024-{i}_val"
    )

# Gather the prediction text files
preds_txt = glob.glob('runs/detect/yolov8-custom-different-augs-image-size-1024-*_val/labels/*.txt')

# Create the save directory if it doesn't exist
os.makedirs(save_path, exist_ok=True)

# Copy the predictions and remove the originals
for i in preds_txt:
    shutil.copy(i, save_path)
    os.remove(i)

print(f"All prediction files have been saved to {save_path}")
```

```python
import pandas as pd
from tqdm.auto import tqdm

# Load the CSV file
input_file = '/kaggle/input/csvthatchdetection/Train.csv'  # Path to the uploaded file
bboxes = pd.read_csv(input_file)

# Initialize dictionary to store worm counts per image and class
images_len = {}

# Process images with bounding boxes
for i in tqdm(bboxes['image_id'].unique()):
    bboxes_temp = bboxes[bboxes['image_id'] == i]
    i = i.split('.')[0]
    
    # Count the number of bounding boxes for each class
    class_counts = bboxes_temp['category_id'].value_counts().to_dict()
    
    # Initialize counts to 0 for each class
    class1_count = class_counts.get(1.0, 0)
    class2_count = class_counts.get(2.0, 0)
    class3_count = class_counts.get(3.0, 0)
    
    # Update the dictionary with counts for each class
    images_len.update({
        f"{i}_class1.jpg": class1_count,
        f"{i}_class2.jpg": class2_count,
        f"{i}_class3.jpg": class3_count,
    })

# Create DataFrame from the dictionary
train_df = pd.DataFrame(images_len.items(), columns=['image_id', 'number_of_worms'])

# Save DataFrame to CSV
output_file = '/kaggle/working/train_modified_new.csv'
train_df.to_csv(output_file, index=False)

print(f"train_modified.csv file has been created at {output_file}")
```

```python
import pandas as pd

# Function to modify image IDs based on class
def modify_image_id(image_id):
    # Replace '_class1.jpg' with '_1.tif'
    new_image_id = image_id.replace('_class1.jpg', '_1.tif')
    # Replace '_class2.jpg' with '_2.tif'
    new_image_id = new_image_id.replace('_class2.jpg', '_2.tif')
    # Replace '_class3.jpg' with '_3.tif'
    new_image_id = new_image_id.replace('_class3.jpg', '_3.tif')
    return new_image_id

# Function to process the CSV file and create a new one
def process_csv(input_file, output_file):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(input_file)
    
    # Modify the Image ID column
    df['image_id'] = df['image_id'].apply(modify_image_id)
    
    # Save the modified DataFrame to a new CSV file
    df.to_csv(output_file, index=False)

# Paths to input and output files
input_csv_file = '/kaggle/working/train_modified_new.csv'
output_csv_file = 'train_OOF.csv'

# Process the CSV file
process_csv(input_csv_file, output_csv_file)

print(f"CSV file '{output_csv_file}' has been created with modified image IDs.")
```

```python
import optuna
import pandas as pd
import os
import numpy as np
from statistics import mean
from ensemble_boxes import soft_nms
from pybboxes import BoundingBox
from joblib import Parallel, delayed
from tqdm.auto import tqdm
import yaml

# Load data
train_labels_df = pd.read_csv('train_modified.csv')
id_label_dict = dict(zip(train_labels_df['image_id'].values, train_labels_df['number_of_worms'].values))

# Define file path for predictions
pred_labels_path = '/kaggle/working'  # Update with your actual predictions path

# MAE Calculation
def mae(y_true, y_pred):
    return np.abs(y_true - y_pred)

# Error Calculation
def return_error(id, pred_label_dict):
    id = id.split('.')[0]
    one_label = id_label_dict.get(f'{id}_1.tif', 0)  # Use .get() to handle missing keys gracefully
    two_label = id_label_dict.get(f'{id}_2.tif', 0)
    three_label = id_label_dict.get(f'{id}_3.tif', 0)
    
    one_pred = pred_label_dict.get(f'{id}_1.tif', 0)
    two_pred = pred_label_dict.get(f'{id}_2.tif', 0)
    three_pred = pred_label_dict.get(f'{id}_3.tif', 0)

    error = float(
        mae(np.array(one_label), np.array(one_pred)) +
        mae(np.array(two_label), np.array(two_pred)) +
        mae(np.array(three_label), np.array(three_pred))
    )
    return error

# Label Generation
def make_labels(id, params):
    id = id.split('.')[0]
    one = 0
    two = 0
    three = 0

    if os.path.exists(f'{pred_labels_path}/{id}_1.tif') and os.path.exists(f'{pred_labels_path}/{id}_2.tif') and os.path.exists(f'{pred_labels_path}/{id}_3.tif'):
        with open(f'{pred_labels_path}/{id}_1.tif') as f1, open(f'{pred_labels_path}/{id}_2.tif') as f2, open(f'{pred_labels_path}/{id}_3.tif') as f3:
            preds_per_line1 = f1.readlines()
            preds_per_line2 = f2.readlines()
            preds_per_line3 = f3.readlines()

            bboxes1 = []
            scores1 = []
            labels1 = []
            bboxes2 = []
            scores2 = []
            labels2 = []
            bboxes3 = []
            scores3 = []
            labels3 = []

            for i in preds_per_line1:
                i = i.split(' ')
                bbox = [float(i[1]), float(i[2]), float(i[3]), float(i[4])]
                bbox = BoundingBox.from_yolo(*bbox, image_size=(1536, 1536))
                bbox = bbox.to_albumentations().values
                bboxes1.append(list(bbox))
                scores1.append(float(i[5]))
                labels1.append(int(i[0]))

            for i in preds_per_line2:
                i = i.split(' ')
                bbox = [float(i[1]), float(i[2]), float(i[3]), float(i[4])]
                bbox = BoundingBox.from_yolo(*bbox, image_size=(1536, 1536))
                bbox = bbox.to_albumentations().values
                bboxes2.append(list(bbox))
                scores2.append(float(i[5]))
                labels2.append(int(i[0]))

            for i in preds_per_line3:
                i = i.split(' ')
                bbox = [float(i[1]), float(i[2]), float(i[3]), float(i[4])]
                bbox = BoundingBox.from_yolo(*bbox, image_size=(1536, 1536))
                bbox = bbox.to_albumentations().values
                bboxes3.append(list(bbox))
                scores3.append(float(i[5]))
                labels3.append(int(i[0]))

            bboxes, scores, labels = soft_nms([bboxes1, bboxes2, bboxes3], [scores1, scores2, scores3], [labels1, labels2, labels3], iou_thr=params['iou_thr'], sigma=params['sigma'], thresh=params['thresh'], method=params['method'])

            for label in labels:
                if label == 1:
                    one += 1
                elif label == 2:
                    two += 1
                elif label == 3:
                    three += 1

    return one, two, three, f"{id}_1.tif", f"{id}_2.tif", f'{id}_3.tif'

# Error Function Class
class ErrorFunc:
    def __init__(self, pred_label_dict):
        self.pred_label_dict = pred_label_dict

    def return_error(self, id):
        id = id.split('.')[0]
        one_label = id_label_dict.get(f'{id}_1.tif', 0)  # Use .get() to handle missing keys gracefully
        two_label = id_label_dict.get(f'{id}_2.tif', 0)
        three_label = id_label_dict.get(f'{id}_3.tif', 0)
        
        one_pred = self.pred_label_dict.get(f'{id}_1.tif', 0)
        two_pred = self.pred_label_dict.get(f'{id}_2.tif', 0)
        three_pred = self.pred_label_dict.get(f'{id}_3.tif', 0)

        error = float(
            mae(np.array(one_label), np.array(one_pred)) +
            mae(np.array(two_label), np.array(two_pred)) +
            mae(np.array(three_label), np.array(three_pred))
        )
        return error

# Objective Function for Optuna
def objective(trial):
    params = {
        'iou_thr': trial.suggest_float('iou_thr', 0.1, 0.7),
        'sigma': trial.suggest_float('sigma', 0.3, 1.0),
        'thresh': trial.suggest_float('thresh', 0.2, 0.6),
        'method': trial.suggest_categorical('method', ['nms', 'linear', 'gaussian']),
    }

    pred = Parallel(n_jobs=12)(delayed(make_labels)(id, params) for id in tqdm(train_labels_df['image_id'].values))
    ids = []
    labels = []
    for i in pred:
        ids.append(i[3])
        ids.append(i[4])
        ids.append(i[5])
        labels.append(i[0])
        labels.append(i[1])
        labels.append(i[2])
    oof = pd.DataFrame({'image_id': ids, 'number_of_worms': labels}, index=None)

    pred_label_dict = dict(zip(oof['image_id'].values, oof['number_of_worms'].values))
    error_fn = ErrorFunc(pred_label_dict)

    error = list(map(error_fn.return_error, train_labels_df['image_id'].values))

    return mean(error)

# Optimize with Optuna
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=2000)

# Save Best Parameters
best_param_save = study.best_params
best_param_save.update({'best_score': study.best_value})
best_param_save.update({'best_trial': study.best_trial.number})
best_param_save.update({'path': pred_labels_path})

with open(f'{pred_labels_path.split("/")[-1]}.yaml', 'w') as f:
    yaml.dump(best_param_save, f)

print("Best parameters saved to YAML file.")
```
